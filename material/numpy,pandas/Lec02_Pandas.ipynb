```python
df = pd.DataFrame({"AAA":[4,5,6,7],
                   "BBB":["남","여","남","여"],
                   "CCC":[100,50,-30,-50]}
                  )
print(df)
print("----------")
def encode(BBB):
  if BBB=="남":
    return 1
  elif BBB=="여":
    return 0
  return -1

print(encode("남"))
print(encode("여"))
print(encode(""))
print(encode("else"))
print("----------")
df['BBB_encoded'] = df[['BBB']].applymap(encode)
print(df)
print("----------")
new_df = df.drop(columns='BBB') # new_df에 복사만 해줌
print(new_df)
print("----------")
df.drop(columns="BBB", inplace=True)  # df에 실제로 적용
print(df)
```

numpy로 0~1 랜덤값을 10개 만들고 
10개의 값을 가지는 r 컬럼의 dataframe을 만들고
5r이라는 r 컬럼에 5배 곱한 컬럼을 새로 만들고
cat이라는 정수값을 하는 컬럼을 만든다.
cat외의 모든 컬럼을 삭제한다.

```python
a = np.random.rand(10)
print(a)
df = pd.DataFrame({"r":a})
print(df)
print("----------")
df['five_r'] = df['r'] * 5
print(df)
print("----------")
df['cat'] = df.five_r.astype(np.int)
print(df)
print("----------")
df.drop(columns=["r","five_r"], inplace=True)
print(df)
```
