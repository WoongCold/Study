```python
a = np.arange(12).reshape((3,4))
print(a)
print(np.sum(a))
a[a>=10] = 0  # 10보다 작은수들의 합
print(a) 
print(np.sum(a))
```

```python
a = np.arange(12).reshape((3,4))
print(a)
mask = (a>5) & (a%2==0) # 5보다 큰 짝수들의 합
print(mask)
print(a*mask) 
print(np.sum(a*mask))
```

```python
a = np.arange(12).reshape((3,4))
print(a)
mask = a%2==0  # 짝수들의 합
print(mask)
print(a*mask)
print(np.sum(a*mask))
```
