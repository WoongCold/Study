```python
sns.histplot(data=titanic,x="age",hue="survived", multiple="fill", shrink=0.7)
plt.savefig("plot3.jpg")

plt.show()
```

```python
plt.subplot(121)
sns.boxplot(data=tips, y="time", x="total_bill")

plt.subplot(122)
sns.violinplot(data=tips, y="time", x="total_bill", order=["Dinner","Lunch"]) # order는 순서 바꿀수가 있다.

plt.tight_layout()
plt.show()
```

```python
plt.subplot(121)
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker") # order는 순서 바꿀수가 있다.

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(121)
sns.boxplot(data=tips, x="day", y="total_bill")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", inner="quartile")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(121)
sns.boxplot(data=tips, x="day", y="total_bill")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", inner="stick")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
np.random.seed(42)

a = np.random.rand(10,8)
print(a[0,0])

plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, vmin=0.2, vmax=0.8)

plt.show()
```
