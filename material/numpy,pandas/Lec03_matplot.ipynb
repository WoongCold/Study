```python
for loc in ["upper left", 
            "upper right", 
            "lower left", 
            "lower right", 
            "center", 
            "center left",
            "center right",
            "right",  
            "best"]:
  fig = plt.figure()
  fig.suptitle(loc)
  plt.plot(x, np.sin(x), label="sin")
  plt.plot(x, np.cos(x), label="cos")
  plt.xlabel("x label")
  plt.ylabel("y label")
  plt.legend(frameon=True, loc=loc)

  plt.show()     
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.cos(x))

ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.cos(x))

ax4 = fig.add_subplot(2,2,4)
ax4.plot(np.sin(x))

plt.show()
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.plot(np.sin(x))
ax1.grid(False)

ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.cos(x))

ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.cos(x))

ax4 = fig.add_subplot(2,2,4)
ax4.plot(np.sin(x))
ax4.grid(False)


plt.show()
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()
fig.suptitle("sin and cos")

ax1 = fig.add_subplot(2,2,1)
ax1.plot(np.sin(x))
ax1.set_title("sin")
ax1.grid(False)

ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.cos(x))
ax2.set_title("cos")

ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.cos(x))
ax3.set_title("cos")

ax4 = fig.add_subplot(2,2,4)
ax4.plot(np.sin(x))
ax4.set_title("sin")
ax4.grid(False)

plt.tight_layout() # 겹치는 것을 제거
plt.show()
```
