```python
fruit_data = [[l,w] for l,w in zip(df.diameter, df.weight)]
fruit_data
```

```python
fruit_target = ['grape']*5000 + ['grapefruit']*5000
print(fruit_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
plt.scatter(orange_diameter, orange_weight)
plt.scatter(grapefruit_diameter, grapefruit_weight)
plt.xlabel('diameter')
plt.ylabel('weight')

plt.show()
```

```python
kn = KNeighborsClassifier()
kn.fit(fruit_data, fruit_target)
```

```python
kn.score(fruit_data, fruit_target)
```

```python
kn.predict([[16,251]])
```

```python
kn.predict([[10.5,171]])
```

```python
kn.predict([[5.0,101]])
```

```python
# diameter와 weight를 비교하여 학습함
for n in range(1,50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수계산
    score = kn.score(fruit_data, fruit_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```

```python
kn = KNeighborsClassifier()
kn.fit(df.drop(columns="name"), df.name)
```

```python
# name을 제외한 모든 속성들과 비교해서 학습함
for n in range(1,5000):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수계산
    score = kn.score(df.drop(columns="name"), df.name)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```
