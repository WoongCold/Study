## alcohol와 sulphates 비교

```python
X1 = red_df['alcohol']
X2 = red_df['sulphates']
Y = red_df['quality']          
```

```python
ax = plt.axes(projection = '3d')
ax.set_xlabel('alcohol')
ax.set_ylabel('sulphates')
ax.set_zlabel('quality')
ax.scatter(X1,X2,Y)

plt.show()
```

```python
X1_data = np.array(X1)
X2_data = np.array(X2)
Y_data = np.array(Y)
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.003
epochs = 10000
```

```python
for i in range(epochs):
    predictY = a1 * X1_data + a2 * X2_data + b
    err = Y_data - predictY
    
    #편미분 연산
    diff_a1 = -(2/len(X1)) * sum(X1_data * err)
    diff_a2 = -(2/len(X2)) * sum(X2_data * err)
    diff_b = -(2/len(Y)) * sum(err)
    
    #가중치와 절편 업데이트
    a1 -= lr * diff_a1
    a2 -= lr * diff_a2
    b -= lr * diff_b
    print(a1,a2,b)
    
    if i%100 == 0:
        print('epoch=%0.4d, a1=%0.4f, a2=%0.4f, b=%0.4f' %(i, a1, a2, b))
print('y = %0.4fx1 + %0.4fx2 + %0.4f' %(a1,a2,b))        
```
