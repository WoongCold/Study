## 선형회귀

```python
Score_data = [[2,81], [4,93], [6,91], [8,97]]
X = [i[0] for i in Score_data]
Y = [i[1] for i in Score_data]
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X,Y)
plt.show()
```

```python
X_data = np.array(X)
Y_data = np.array(Y)
```

```python
a = 0
b = 0
```

```python
lr = 0.03
epochs = 1000
```

```python
for i in range(epochs):
    predict_y = a * X_data + b
    err = Y_data - predict_y
    
    #편미분 연산
    a_dif = -(1/len(X_data)) * sum(X_data * err)
    b_dif = -(1/len(X_data)) * sum(err)
    
    #가중치와 절편 업데이트
    a = a - lr * a_dif
    b = b - lr * b_dif
    
    if i%100 == 0:
        print('epoch=%f, 기울기=%0.4f  절편=%0.4f' %(i, a, b))
```

```python
def top(length, mean_length, weight, mean_weight):
    diffvalue = 0
    
    for i in range(len(length)):
        diffvalue += (length[i] - mean_length) * (weight[i] - mean_weight)
    return diffvalue

diffnum = top(length, mean_length, weight, mean_weight)

print('분모: ', diffx)
print('분자: ', diffnum)

a = diffnum / diffx
b = mean_weight - (mean_length*a)

print('기울기 a = ', a)
print('y절편 b = ', b)
```

## 다중 선형 회귀

```python
input_data = [[2,0,81], [4,4,93], [6,2,91], [8,3,97]]
X1 = [i[0] for i in input_data]
X2 = [i[1] for i in input_data]
Y = [i[2] for i in input_data]
```

```python
ax = plt.axes(projection = '3d')
ax.set_xlabel('Study times')
ax.set_ylabel('Private_class')
ax.set_zlabel('Score')
ax.scatter(X1,X2,Y)

plt.show()
```

```python
X1_data = np.array(X1)
X2_data = np.array(X2)
Y_data = np.array(Y)
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.002
epochs = 10000
```

```python
for i in range(epochs):
    predictY = a1 * X1_data + a2 * X2_data + b
    err = Y_data - predictY
    
    #편미분 연산
    diff_a1 = -(2/len(X1)) * sum(X1_data * err)
    diff_a2 = -(2/len(X2)) * sum(X2_data * err)
    diff_b = -(2/len(Y)) * sum(err)
    
    #가중치와 절편 업데이트
    a1 -= lr * diff_a1
    a2 -= lr * diff_a2
    b -= lr * diff_b
    print(a1,a2,b)
    
    if i%100 == 0:
        print('epoch=%0.4d, a1=%0.4f, a2=%0.4f, b=%0.4f' %(i, a1, a2, b))
print('y = %0.4fx1 + %0.4fx2 + %0.4f' %(a1,a2,b))                
```
